```python
import os

from streetview import search_panoramas, get_panorama_meta, get_streetview, get_panorama

GOOGLE_MAPS_API_KEY = os.getenv("GOOGLE_MAPS_API_KEY")
```

```python
panos = search_panoramas(lat=41.8982208, lon=12.4764804)
panos
```

```python
pano_id = panos[1].pano_id
```

```python
meta = get_panorama_meta(pano_id=pano_id, api_key=GOOGLE_MAPS_API_KEY)

print(meta)
```

```python
image = get_streetview(
    pano_id=pano_id,
    api_key=GOOGLE_MAPS_API_KEY,
)
image
```

```python
img_panorama = get_panorama(pano_id, zoom=1)
```
